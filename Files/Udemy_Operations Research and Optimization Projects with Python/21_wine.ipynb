```python
import pulp as pl

# Define the data
data = {
    'type': ['red']*10 + ['white']*10 + ['other']*10,  # Types for each wine
    'rating': [85, 90, 88, 87, 89, 90, 91, 92, 93, 88,  # Ratings for each wine
               86, 87, 85, 88, 89, 90, 91, 92, 93, 94,
               82, 84, 86, 88, 90, 92, 94, 96, 98, 99],
    'price': [30, 45, 20, 25, 30, 35, 40, 45, 50, 55,  # Prices for each wine
              20, 25, 30, 35, 40, 45, 50, 55, 60, 65,
              15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
}

# Create the optimization problem
prob = pl.LpProblem("Wine_Selection", pl.LpMaximize)

# Decision variables
x = pl.LpVariable.dicts("Select", list(range(30)), cat='Binary')

# Objective Function: Maximize the sum of ratings of the selected wines
prob += pl.lpSum([data['rating'][i] * x[i] for i in range(30)])

# Constraints
# Select 3 red wines
prob += pl.lpSum([x[i] for i in range(10)]) == 3

# Select 1 white wine
prob += pl.lpSum([x[i] for i in range(10, 20)]) == 1

# Select 2 wines from other types
prob += pl.lpSum([x[i] for i in range(20, 30)]) == 2

# Solve the problem
prob.solve()

# Print the results
selected_wines = [i for i in range(30) if pl.value(x[i]) == 1]
print("Selected Wines Indices:", selected_wines)
print("Selected Wines Details:")
for i in selected_wines:
    print(f"Wine {i} - Type: {data['type'][i]}, Rating: {data['rating'][i]}, Price: {data['price'][i]}")
```
