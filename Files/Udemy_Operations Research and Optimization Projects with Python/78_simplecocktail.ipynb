```python
import pulp

# Create a LP problem
model = pulp.LpProblem("Cocktail_Optimization", pulp.LpMaximize)

# Decision Variables
cocktails = ['Margarita', 'Mojito', 'Cosmopolitan']
include_cocktail = pulp.LpVariable.dicts("Include", cocktails, cat='Binary')

# Costs and Satisfaction Scores (Example data)
costs = {'Margarita': 4, 'Mojito': 3.5, 'Cosmopolitan': 5}
satisfaction = {'Margarita': 80, 'Mojito': 70, 'Cosmopolitan': 85}

# Parameters
budget = 10  # Total budget

# Objective Function: Maximize total satisfaction
model += pulp.lpSum([include_cocktail[c] * satisfaction[c] for c in cocktails]), "Total_Satisfaction"

# Constraint: Stay within budget
model += pulp.lpSum([include_cocktail[c] * costs[c] for c in cocktails]) <= budget, "Budget_Constraint"

# Solve the problem
model.solve()

# Output the results
print("Status:", pulp.LpStatus[model.status])
for c in cocktails:
    print(f"{c} included in the menu: {'Yes' if include_cocktail[c].value() == 1 else 'No'}")

print("Total Cost:", pulp.value(model.objective))
```
