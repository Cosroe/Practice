Let $y=e^{-x/10}sin(x)$. Consider 10000 $x$ intervals in the range [0,10].

1. Plot the function $y$ vs. $x$ in the range [0,10].

```python
# 1
x = np.linspace(0,10,10000)
y = np.exp(-x/10) * np.sin(x)
plt.plot(x,y)
```

The power emitted by a blackbody is $P=A\sigma\epsilon T^4$. After measuring the temperature of a star you find that $T(t)=T_{0}\frac{1}{1+e^{-kt}}$. Plot the total energy emitted by the star as a function of time using the fact that $E(t)=\int_{t^{'}=0}^{t^{'}=t} P(t^{'}) \,dt$.

First let's rearrange the equations:
 $$\frac{P}{A \sigma \epsilon T_{0}^{4}} = (\frac{1}{1+e^{-kt}})^{4}$$

```python
kt = np.linspace(0,3,100)
P = (1/(1+np.exp(-kt)))**4
```

```python
plt.plot(kt,P)
```

```python
E = np.cumsum(P) * (kt[1]-kt[0])
```

```python
plt.plot(kt,E)
plt.xlabel('$kt$')
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$')
```

An electric field is given by 
$$\overrightarrow{E}(z,t) = E_{0}cos(z-t)\hat{x} + 2E_{0}cos(z-t+ \frac{\pi}{2})\hat{y}$$

1. Find the magnetic field for $\forall z \in [0,4\pi]$ and $t \in [0,10]$ using $\overrightarrow{B}(z,t)=\hat(z) \times \overrightarrow{E}(z,t)$.
2. Compute the Poynting vector $\overrightarrow{S} = \overrightarrow{E} \times \overrightarrow{B}$ for all $z$ and $t$.

```python
#Q1
z = np.linspace(0, 4*np.pi, 100)
t = np.linspace(0, 10, 100)

tv, zv = np.meshgrid(t,z)

Ex = np.cos(zv-tv)
Ey = 2*np.cos(zv-tv + np.pi/2)
Ez = 0*tv
```

```python
plt.plot(t, Ex[0])
```

```python
E = np.array([Ex, Ey, Ez])
E
```

```python
E = np.swapaxes(E,0,-1)
```

```python
E
```

```python
B = np.cross(np.array([0,0,1]), E)
```

```python
B = np.swapaxes(B,0,-1)
```

```python
Bx, By, Bz = B
```

```python
B
```

```python
plt.plot(t, Ey[0])
plt.plot(t,Bx[0])
```
